### Criando Gráficos

Claro! Vamos lá:

Criar gráficos é o processo de visualizar dados de forma gráfica para facilitar a compreensão e a análise de informações. Aqui estão os principais pontos sobre a criação de gráficos:

1. **Escolha do Tipo de Gráfico**: O primeiro passo é escolher o tipo de gráfico adequado aos dados e ao tipo de análise que você deseja realizar. Exemplos comuns incluem gráficos de barras, gráficos de linhas, gráficos de dispersão e gráficos de pizza.

2. **Bibliotecas de Visualização**: Em Python, existem várias bibliotecas populares para criar gráficos, como Matplotlib, Seaborn e Plotly. Cada uma oferece diferentes funcionalidades e estilos de visualização.

3. **Preparação dos Dados**: Antes de criar um gráfico, os dados devem estar organizados e preparados. Isso pode incluir a seleção de colunas relevantes, filtragem de dados e cálculo de estatísticas necessárias.

4. **Criação do Gráfico**:
    - **Matplotlib**: É uma biblioteca básica e versátil. Por exemplo, para criar um gráfico de linha:
      ```python
      import matplotlib.pyplot as plt

      plt.plot(x, y)
      plt.xlabel('Eixo X')
      plt.ylabel('Eixo Y')
      plt.title('Título do Gráfico')
      plt.show()
      ```
    - **Seaborn**: É uma biblioteca baseada no Matplotlib que oferece gráficos mais sofisticados com menos código. Por exemplo, para criar um gráfico de barras:
      ```python
      import seaborn as sns

      sns.barplot(x='categoria', y='valor', data=meu_dataframe)
      plt.show()
      ```
    - **Plotly**: Oferece gráficos interativos. Por exemplo, para criar um gráfico de dispersão:
      ```python
      import plotly.express as px

      fig = px.scatter(meu_dataframe, x='coluna_x', y='coluna_y', title='Título do Gráfico')
      fig.show()
      ```

5. **Personalização do Gráfico**: Após criar o gráfico, é possível personalizá-lo ajustando cores, rótulos, títulos, legendas e outros elementos para melhorar a clareza e a estética da visualização.

6. **Exibição e Salvamento**: Finalmente, você pode exibir o gráfico na tela ou salvá-lo em um arquivo para uso posterior. Em Matplotlib, por exemplo, você pode salvar o gráfico usando `plt.savefig('nome_do_arquivo.png')`.

Em resumo, criar gráficos envolve escolher o tipo certo de gráfico, preparar os dados, usar uma biblioteca de visualização para gerar o gráfico e personalizá-lo conforme necessário para comunicar efetivamente as informações.

````python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('netflix.csv')
df = df.drop(df.columns[0], axis=1)
````

````python
df.head()


````

````python
eixo_x = df['nível_de_classificação'].value_counts().keys()
eixo_y = df['nível_de_classificação'].value_counts().values
````

````python
# Renomear as colunas para português (opcional):
df.columns = ['título', 'classificação', 'nível_de_classificação', 'descrição_da_classificação', 'ano_de_lançamento', 'pontuação_de_classificação_do_usuário']
````

````python
# Remover valores NaN da coluna 'pontuação_de_classificação_do_usuário' antes de plotar
df_cleaned = df['nível_de_classificação'].dropna()

# Criar um histograma para visualizar a distribuição das pontuações de classificação dos usuários
plt.hist(df_cleaned, bins=20, color='skyblue', edgecolor='black')

# Adicionar um título ao gráfico
plt.title('Distribuição das Pontuações de Classificação')

# Adicionar um rótulo ao eixo x
plt.xlabel('Pontuação de Classificação')

# Adicionar um rótulo ao eixo y
plt.ylabel('Frequência')

# Adicionar uma grade ao fundo do gráfico
plt.grid(True)

# Mostrar o gráfico
plt.show()
````

````python
top_10_scores = df['nível_de_classificação'].value_counts().head(10)

plt.figure(figsize=(10,6))
top_10_scores.plot(kind='bar', color='skyblue')
plt.title('Top 10 Pontuações de Classificação')
plt.xlabel('Pontuação de Classificação')
plt.ylabel('Frequência')
plt.xticks(rotation=45)
plt.show()
````

````python
plt.figure(figsize=(10,6))
plt.scatter(df['descrição_da_classificação'], df['nível_de_classificação'], color='skyblue')
plt.title('Pontuação de Classificação do Usuário por Ano de Lançamento')
plt.xlabel('Ano de Lançamento')
plt.ylabel('Pontuação de Classificação')
plt.grid(True)
plt.show()
````
